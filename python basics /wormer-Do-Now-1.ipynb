# 4) Given the following code, what would the output be for n = 0, 10, -1, 3 and-5?

For n = 0 the output must be F and H, because n is not bigger than 0, so the first if statement is not fulfilled. But the second and third if statements are fulfilled, so python prints F and H.


```python
n = 0

print n
if n > 0:
    print "A"
    if n >= 3:
        print "B"
        print "C"
    elif n > 2:
        print "D"
    else:
        print "E"
elif 1 > n > -2:
    print "F"
else:
    print "G"
if n < 5:
    print "H"
```
